```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

dist = np.random.normal(loc=10,size=100000)
dist = dist[dist >= 0] 
sns.distplot(dist)
```

```python
dist1 = np.power(dist, 6)
sns.distplot(dist1)
```

```python
dist = np.random.normal(loc=1000,scale=100, size=100000)
dist = dist[dist >= 0] 
dist1 = np.power(dist, 1/6)
sns.distplot(dist1)
```

```python
dist = np.random.normal(loc=10,size=100000)
dist = dist[dist >= 0] 
dist1 = np.power(3, dist)
sns.distplot(dist1)
```
